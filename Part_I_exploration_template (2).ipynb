### Data Cleaning

#### 1. Duplicated records

Further assesing the dataset to make sure the duplicates are actually duplicated and not just mistakes.

```python
# Checking the duplicates using more than one column.
duplicateRows = loan_sub[loan_sub.duplicated(['ListingNumber', 'Occupation', 'BorrowerState'])]
```

```python
duplicateRows
```

>Listing out the duplicated rows to do a manual assessment.

```python
loan_sub[loan_sub['ListingNumber'].isin(duplicateRows.ListingNumber)].sort_values(by=['ListingNumber'])
```

> It is now obvious that the duplicate enteries have been duplicated on all columns except the ProsperScore column. With respect to the similarities in all the other columns, the difference in that one column can be ignored and therefore the duplicated records will be dropped.

```python
loan_sub.drop_duplicates(subset=['ListingNumber'], inplace=True)
```

```python
loan_sub.info()
```

#### 2. Wrong data types
Changing:
- ListingNumber from int64 to object
- LoanOriginationDate from object to datetime

```python
# To object
loan_sub['ListingNumber'] = loan_sub['ListingNumber'].astype(str)

# To datetime
loan_sub['LoanOriginationDate'] = pd.to_datetime(loan_sub['LoanOriginationDate'])
```
